```python
import numpy as np
import pandas as pd
from sklearn import tree

df = pd.read_csv("DecisionTreesClassificationDataSet.csv")
```

```python
df
```

```python
mapping1={"Y":1,"N":0}
df['IseAlindi'] = df['IseAlindi'].map(mapping1)
df['SuanCalisiyor?'] = df['SuanCalisiyor?'].map(mapping1)
df['Top10 Universite?'] = df['Top10 Universite?'].map(mapping1)
df['StajBizdeYaptimi?'] = df['StajBizdeYaptimi?'].map(mapping1)
mapping2={"BS":0,"MS":1,"PhD":2}
df['Egitim Seviyesi'] = df['Egitim Seviyesi'].map(mapping2)
```

```python
df
```

```python
y=df["IseAlindi"]
X=df.drop(["IseAlindi"],axis=1)
```

```python
model=tree.DecisionTreeClassifier()
model_fit=model.fit(X,y)
```

```python
print (model_fit.predict([[5, 1, 3, 0, 0, 0]]))
```

```python
print (model_fit.predict([[2, 0, 7, 0, 1, 0]]))
```

```python
print (model_fit.predict([[20, 0, 5, 1, 1, 1]]))
```
